# Finding out the Inter Quantile range and then with help of IQR method elemenating the outliers in the data set

```python
num_feature=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']
# Sacling the numerical values with logarithmic  function.
for feature in num_feature:
    df[feature]=np.log(df[feature])

def iqr_feature(num_feature, df):
    dict = {}
    max_lmt = []
    min_lmt=[]
    for feature in num_feature:
        q1 = df[feature].quantile(0.25)
        q2 = df[feature].quantile(0.75)
        IQR = q2 - q1
    
        max_limit = q2 + (1.5 * IQR)
        max_lmt.append(max_limit)
        min_limit = q1 - (1.5 * IQR) 
        min_lmt.append(min_limit)

    return max_lmt, min_lmt
    
max_lmt, min_lmt= iqr_feature(num_feature, df)
```
